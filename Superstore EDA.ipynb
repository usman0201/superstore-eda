```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel('Superstore.xlsx')
```

```python
df
```

```python
pd.set_option('display.max_rows',None)
```

# Correlation

```python
df.corr()
```

```python
sns.heatmap(df.corr(),annot=True)
```

# Profit on the basis of class and segment

```python
sns.barplot(x='Ship Mode',y='Profit',data=df,hue='Segment')
```

# Hightest and Lowest sales on basis of cities

```python
df['City'].value_counts().nlargest(10).plot(kind='bar',color='pink')
```

# Most sold Category

```python
df['Category'].value_counts().nlargest(10).plot(kind='bar',color='green')
```

# Most and least sold item

```python
df['Sub-Category'].value_counts()
```

```python
df['Sub-Category'].value_counts().nlargest(5).plot(kind='bar',color='red')
```

# Profit on the basis of Region

```python
sns.barplot('Region','Profit',data=df)
```

# Region wise transactions

```python
df['Region'].value_counts().plot(kind='pie')
```

```python
data=df.sort_values('Profit',ascending=False).head(10)
```

# Most profitable cities and city with highest loss

```python
fig_dims = (12,10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x='City',y='Profit',data=data)
```

```python
data=df.sort_values('Profit',ascending=True).head(10)
```

```python
fig_dims = (12,10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x='City',y='Profit',data=data)
```

# Profit and sale scatter plot on the basis of region

```python
fig_dims = (8,8)
fig, ax = plt.subplots(figsize=fig_dims)
sns.scatterplot(x='Sales',y='Profit',data=df,hue='Region')
```
